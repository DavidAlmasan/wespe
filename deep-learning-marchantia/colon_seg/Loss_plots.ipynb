```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pretrained_losses = np.load('losses_pretrained.npy')
not_pretrained_losses = np.load('losses_not_pretrained.npy')
```

```python
plt.plot(1 - pretrained_losses)
plt.plot(1 - not_pretrained_losses)
plt.xlabel('epoch')
plt.ylabel('Dice Coefficient')
plt.legend(['pretrained','from scratch'])
plt.savefig('loss_plot.png', dpi = 100)
```

```python
losses_vgg11 = np.load('./results/diffencoders/losses_pretrained_vgg11.npy')
losses_vgg16 = np.load('./results/diffencoders/losses_pretrained_vgg16.npy')
losses_resnet34 = np.load('./results/diffencoders/losses_pretrained_resnet34_2.npy')
```
